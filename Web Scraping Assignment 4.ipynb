```python
title=[]
year=[]
for i in range(0,len(b1)):
    title.append(b1[i][0])
    year.append(b1[i][1])
    
    
```

```python
#Scraping Genre
gen=[]
c=driver.find_elements(By.XPATH,"//span[@class='genre']")
for i in c:
    gen.append(i.text)
```

```python
#Scraping Runtime
run=[]
d=driver.find_elements(By.XPATH,"//span[@class='runtime']")
for i in d:
    run.append(i.text)
```

```python
#Scraping Ratings
rate=[]
e=driver.find_elements(By.XPATH,"//div[@class='ipl-rating-star small']")
for i in e:
    rate.append(i.text)
```

```python
#Scraping votes
vote=[]
f=driver.find_elements("name","nv")
for i in f:
    vote.append(i.text)
```

```python
#Creating Dataframe
data=pd.DataFrame({"Name":title,"Year Span":year,"Genre":gen,"Runtime":run,"Ratings":rate,"Votes":vote})
data
```
